```python
from ultralytics import YOLO
from PIL import Image
import cv2
import numpy
```

```python
# Load a pretrained YOLOv8n model
model = YOLO('yolov8n-face.pt')

# Run inference on an image
image_path = 'image_pulp_test.jpg'
results = model('image_pulp_test.jpg', save = True)  # results list

# View results
for r in results:
    print(r.boxes)  # print the Boxes object containing the detection bounding boxes
```

```python
from PIL import Image
import matplotlib.pyplot as plt


# Assuming 'image_path' and 'results' are defined as in your previous code

# Load the original image
original_image = Image.open(image_path)

# Get the bounding box coordinates from the first detection result
xyxy_tensor = results[0].boxes.xyxy
xyxy_list = xyxy_tensor.numpy().astype(int)

# Loop through each set of coordinates
for coord in xyxy_list:
    # Convert the bounding box coordinates to a tuple
    coord_tuple = tuple(coord)

    # Crop the region of interest (ROI) from the original image
    cropped_image = original_image.crop(coord_tuple)

    plt.imshow(cropped_image)
    plt.axis('off')  # Turn off axis labels
    plt.show()
```
