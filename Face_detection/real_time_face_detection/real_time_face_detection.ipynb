### Découper la vidéo en frame

```python
import argparse
import time
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
from ultralytics import YOLO
from PIL import Image
```

```python
model = YOLO('../First_test/yolov8n-face.pt')
```

Tester en extrayant les visages

```python
video_path = "PF_milkshake.mp4"

capture = cv2.VideoCapture(video_path)
frameNr = 0
os.makedirs("output", exist_ok=True)

while True:
    success, frame = capture.read()
    fps = int(capture.get(cv2.CAP_PROP_FPS))
    
    if success:
        #Thread(target=predict, args=frame).start()
        yolov_result = model(frame)
        xyxy_tensor = yolov_result[0].boxes.xyxy
        xyxy_list = xyxy_tensor.numpy().astype(int)
        
        for coord in xyxy_list:
            # Convert the bounding box coordinates to a tuple
            coord_tuple = tuple(coord)

            # Crop the region of interest (ROI) from the original image
            roi = frame[coord_tuple[1]:coord_tuple[3], coord_tuple[0]:coord_tuple[2]]
        
        if frameNr % (fps) == 0:
            cv2.imwrite(f'output/frame_{frameNr}.jpg', roi)

    else:
        break

    frameNr = frameNr+1

capture.release()
```
